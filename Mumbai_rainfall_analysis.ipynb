```python
#importing all neccessary libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import csv
import seaborn as sns 


#Reading the CSV file and assigning it to df
df = pd.read_csv('mumbai-monthly-rains.csv') 

#assigning a new variable to a new data frame with the final 10 rows
df_new = df.tail(10)


#assigning variables for plotting
year=df['Year']
rainfall_in_march=df['Mar']
year_new = df_new['Year']
Total = df_new['Total']


#defining the function for line plot
def line_plot_rainfall(year, rainfall_in_march): 
    """creates a line plot with year on x axis and 
    rainfall in March on y axis"""


    #create figure and plot the line graph
    plt.figure(figsize = (6,4)) 
    plt.plot(year, rainfall_in_march, color='black')  
    
    #setting the title
    plt.title("Rainfall Received in the month of March\nover the Years 1901-2021 in Mumbai", fontsize=10, color='darkblue', loc='center')
    
    #setting the x ticks and y ticks
    plt.xticks(np.arange(1900, 2022, 10),rotation= 45,fontsize = 10)
    plt.yticks(size = 10)                

     # Adjust layout for better readability
    plt.tight_layout() 

    #setting x and y label
    plt.xlabel('Years(1901-2021)', fontsize = 11, color= 'teal') 
    plt.ylabel('Rainfall in March\n(in mm)', fontsize = 11, color = 'teal')

    #Adding grid and displaying the plot
    plt.grid(True) 
    plt.show()  
    return
    
# Calling the function
line_plot_rainfall(year, rainfall_in_march)  # Calling the function

#printing the mean, median and standard deviation of rainfall in march
print("rainfall in March:median=",rainfall_in_march.median())  
print("rainfall in March:mean=",rainfall_in_march.mean()) 
print("rainfallin March:standard deviation=",rainfall_in_march.std()) 

#printing skewness and kurtosis
skew_kurt_of_rainfall_in_march=df[['Mar']].copy()
print("skewness of rainfall in march",skew_kurt_of_rainfall_in_march.skew())
print("kurtosis of rainfall in march",skew_kurt_of_rainfall_in_march.kurtosis()) 
print("numerical statistics=",df['Mar'].describe())



# defining a function for plottimg bar graph
def plot_bar(year_new, Total): 


    #setting figuresize and making the plot
    plt.figure(figsize = (6,4))  
    plt.bar(year_new, Total, color = 'brown') 
    
    #setting xlabel and ylabel
    plt.xlabel('Years(2012-2021)', color = 'teal') 
    plt.ylabel('Total amount of rainfall received\nin Mumbai per year\n(in mm)', fontsize = 12, color = 'teal') 
    
    #setting the title
    plt.title('Rainfall received from 2012 to 2021', color = 'darkblue') 
    
    #setting xticks and yticks
    plt.xticks(np.arange(2012, 2022, 1),rotation=45, fontsize = 10) 
    plt.yticks(fontsize = 10) 

    #displaying the plot
    plt.show()
    return


#calling the function
plot_bar(year_new, Total)

#printing the mean, median and standard deviation of total rainfall
print("Total rainfall:median=",Total.median())
print("Total rainfall:mean=",Total.mean()) 
print("Total rainfall:standard deviation=",Total.std())


Total_rainfall=df[['Total']].copy()

#printing the skewness, kurtosis and numerical statistics of total rainfall
print("skewness of Total rainfall",Total_rainfall.skew()) #finding skewness of total rainfall
print("kurtosis of Total rainfall",Total_rainfall.kurtosis()) #finding kurtosis of total rainfall
print("numerical statistics=",df_new['Total'].describe())



#defining a function for coprrelation heat map
def plot_correlation(df, method):

    #setting the figuresize
    plt.figure(figsize=(8,4)) 

    #masking the upper triangle because it is symmetric
    mask = np.triu(np.ones_like(df.corr())) 

    #plotting the heatmap
    sns.heatmap(df.corr(method=method), vmin=-1, vmax=1, cmap='Blues', annot=True, mask=mask, annot_kws={"fontsize":7})

    #setting the title
    plt.title('Heat map of Annual Rainfall')
    
    #displaying the heatmap
    plt.show()
    return


#pearson correlation heatmap
plot_correlation(df, 'pearson')

#printing out correlation and numerical statistics
print("correlation matrix=",df.corr())
print("numerical statistics=",df.describe()) 


```
